## Did the median value change?

```python
# Making a copy of dataframe
df1 = df.copy()

# Finding median value for "Engine Cylinders"
engine_cylinders_median = df1["Engine Cylinders"].median()

# Display result
print(engine_cylinders_median)
```

```python
# Determining number of missing values in "Engine Cylinders"
df["Engine Cylinders"].isna().sum()
```

```python
# Finding the most frequent value for "Engine Cylinders"
engine_cylinders_mode = df1["Engine Cylinders"].mode()

# Display result
print(engine_cylinders_mode)
```

```python
# Filling in any missing values in "Engine Cylinders"
df2 = df.copy()

# Filling in NA values with median
df2["Engine Cylinders"] = df2["Engine Cylinders"].fillna(engine_cylinders_mode)
```

```python
# Calculating new median value for "Engine Cylinders"
new_engine_cylinders_median = df2["Engine Cylinders"].median()

# Display result
print(new_engine_cylinders_median)
```

```python
# Display result
print("After using fillna to change the median of ""Engine Cylinders"", the median value changed:", engine_cylinders_median != new_engine_cylinders_median)
```

# Question 7

* Select all the "Lotus" cars from the dataset.
* Select only columns "Engine HP", "Engine Cylinders".
* Now drop all duplicated rows using drop_duplicates method (you should get a dataframe with 9 rows).
* Get the underlying NumPy array. Let's call it X.
* Compute matrix-matrix multiplication between the transpose of X and X. To get the transpose, use X.T. Let's call the result XTX.
* Invert XTX.
* Create an array y with values [1100, 800, 750, 850, 1300, 1000, 1000, 1300, 800].
* Multiply the inverse of XTX with the transpose of X, and then multiply the result by y. Call the result w.


## What's the value of the first element of w?

```python
# Selecing all the Lotus cars from the dataset
lotus_df = df[df["Make"] == "Lotus"]
lotus_df
```

```python
# Select only columns "Engine HP", "Engine Cylinders".

columns = ["Engine HP", "Engine Cylinders"]

# Making a dataframe of just "Engine HP" and "Engine Cylinders"
lotus_df = lotus_df[columns]
lotus_df
```

```python
# Dropping duplicates in lotus_df, dataframe should just have 9 rows
lotus_df = lotus_df.drop_duplicates()

# Confirming dataframe just has 9 rows
print("The Lotus dataframe now has 9 rows:", lotus_df.shape[0] == 9)
```

```python
# Getting the undelying numpy array
X = lotus_df.to_numpy()
X
```

```python
# Computing the matrix-matrix multiplication between the transpose of X and X. 
XTX = X.T.dot(X)
XTX
```

```python
# Creating an array y with [1100, 800, 750, 850, 1300, 1000, 1000, 1300, 800]
y = np.array([1100, 800, 750, 850, 1300, 1000, 1000, 1300, 800])
y
```

```python
# Multiply the inverse of XTX with the transpose of X, multiply the result by y, call the result w
from numpy.linalg import inv
XTX_inverse = inv(XTX)

# Saving the result in w
w = XTX_inverse.dot(X.T).dot(y)
```

```python
# Getting the first value of w
print("The first value of w is:", w[0])
```
